```python
import pandas as pd

import quantstats as qs
```

```python
dados_b3 = pd.read_csv("/Users/pedroafmelo/Documents/projetos/POO_finance/data/b3_companies.csv.gz")

dados_b3
```

```python
dados_empresas = pd.read_csv("/Users/pedroafmelo/Documents/projetos/POO_finance/data/dados_empresas.csv")

dados_empresas["retorno"] = dados_empresas.groupby("ticker")["preco_fechamento_ajustado"].pct_change()
dados_empresas["retorno"] = dados_empresas.groupby("ticker")["retorno"].shift(-1)

dados_empresas
```

```python
dados_empresas = dados_empresas.query("volume_negociado > 1000000")
```

```python
dados_empresas
```

```python
rentabilidade_carteira = dados_empresas.groupby("data")["retorno"].mean()
```

```python
rentabilidade_carteira = rentabilidade_carteira.to_frame()
```

```python
rentabilidade_carteira['estrategia_dy'] = (rentabilidade_carteira['retorno'] + 1).cumprod() - 1

rentabilidade_por_carteiras = rentabilidade_carteira.shift(1)

rentabilidade_por_carteiras = rentabilidade_por_carteiras.dropna()
```

```python
rentabilidade_carteira
```

```python
ibov = pd.read_csv('/Users/pedroafmelo/Documents/projetos/POO_finance/data/ibov.csv')

retornos_ibov = ibov['fechamento'].pct_change().dropna()
retorno_acum_ibov = (1 + retornos_ibov).cumprod() - 1
rentabilidade_por_carteiras['Ibovespa'] = retorno_acum_ibov.values
rentabilidade_por_carteiras = rentabilidade_por_carteiras.drop('retorno', axis = 1)
```

```python
qs.extend_pandas()
rentabilidade_por_carteiras.index = pd.to_datetime(rentabilidade_por_carteiras.index)

rentabilidade_por_carteiras['estrategia_dy'].plot_monthly_heatmap()
rentabilidade_por_carteiras['Ibovespa'].plot_monthly_heatmap()
```
